## Phase 2 - Data Cleaning
We use the BudgetWise Dirty dataset as our primary working dataset.
Cleaning steps:
1. Remove duplicates
2. Fix data types (amount, date)
3. Handle missing values
4. Drop the notes column
5. Save the cleaned dataset

```python
# Cell 7 - Remove Duplicates
print(f"Rows before removing duplicates: {bw_dirty.shape[0]:,}")

bw = bw_dirty.drop_duplicates()

print(f"Rows after removing duplicates: {bw.shape[0]:,}")
print(f"Duplicates removed: {bw_dirty.shape[0] - bw.shape[0]:,}")
```

```python
# Cell 8 - Fix Data Types
# Fix amount - remove any currency symbols and convert to float
bw['amount'] = bw['amount'].str.replace(',', '').str.replace('$', '').str.strip()
bw['amount'] = pd.to_numeric(bw['amount'], errors='coerce')

# Fix date - convert string to proper datetime
bw['date'] = pd.to_datetime(bw['date'], errors='coerce')

# Verify the fix
print("Data types after fixing:")
print(bw[['date', 'amount']].dtypes)
print()
print(f"Amount - sample values: {bw['amount'].dropna().head().tolist()}")
print(f"Date - sample values: {bw['date'].dropna().head().tolist()}")
print()
print(f"Amount nulls after conversion: {bw['amount'].isnull().sum()}")
print(f"Date nulls after conversion: {bw['date'].isnull().sum()}")
```

```python
# Cell 9 - Investigate date and amount issues
# Look at raw date values that failed to parse
print("Sample raw date values from dirty dataset:")
print(bw_dirty['date'].dropna().unique()[:20])
print()

# Look at raw amount values that failed to parse
print("Sample raw amount values that became null:")
null_amounts = bw_dirty[pd.to_numeric(
    bw_dirty['amount'].str.replace(',', '').str.replace('$', '').str.strip(),
    errors='coerce'
).isnull()]['amount'].dropna()
print(null_amounts.unique()[:20])
```

```python
# Cell 10 - Fix Data Types Properly

# Reload bw from scratch to ensure it's in string format
bw = bw_dirty.drop_duplicates().copy()

# Fix amount - strip both $ and rupee symbol
bw['amount'] = (bw['amount']
                .str.replace('₹', '', regex=False)
                .str.replace('$', '', regex=False)
                .str.replace(',', '', regex=False)
                .str.strip())
bw['amount'] = pd.to_numeric(bw['amount'], errors='coerce')

# Fix date - pandas handles mixed formats with dayfirst=False
bw['date'] = pd.to_datetime(bw['date'], errors='coerce', dayfirst=False)

# Verify
print("Data types after fixing:")
print(bw[['date', 'amount']].dtypes)
print()
print(f"Amount nulls remaining: {bw['amount'].isnull().sum()}")
print(f"Date nulls remaining: {bw['date'].isnull().sum()}")
print()
print(f"Amount sample: {bw['amount'].dropna().head().tolist()}")
print(f"Date sample: {bw['date'].dropna().head().tolist()}")
```

```python
# Cell 11 - Investigate date nulls
# How many dates were null in the original dirty dataset?
print(f"Original null dates in dirty dataset: {bw_dirty['date'].isnull().sum()}")
print(f"Total rows in dirty dataset: {bw_dirty.shape[0]}")
print()

# What do the non-null raw date values look like?
raw_dates = bw_dirty['date'].dropna()
print(f"Non-null date count: {len(raw_dates)}")
print()

# After parsing, how many became null?
parsed = pd.to_datetime(raw_dates, errors='coerce', dayfirst=False)
failed = parsed.isnull().sum()
print(f"Dates that failed to parse: {failed}")
print()

# Show samples of dates that failed to parse
failed_dates = raw_dates[parsed.isnull()]
print(f"Sample failed date values:")
print(failed_dates.unique()[:20])
```

```python
# Cell 12 - Fix dates with multiple format parser
from dateutil import parser as dateutil_parser

def parse_date_flexible(date_str):
    if pd.isnull(date_str):
        return pd.NaT
    try:
        return dateutil_parser.parse(str(date_str), dayfirst=False)
    except:
        return pd.NaT

# Apply flexible parser
print("Parsing dates with flexible parser...")
bw['date'] = bw_dirty.drop_duplicates()['date'].apply(parse_date_flexible)

# Verify
print(f"Date nulls remaining: {bw['date'].isnull().sum()}")
print(f"Successfully parsed: {bw['date'].notna().sum():,}")
print()
print(f"Sample parsed dates:")
print(bw['date'].dropna().head(10).tolist())
```

```python
# Cell 13 - Handle Missing Values

# Drop notes column - 10% missing, free text, not useful for modeling
bw = bw.drop(columns=['notes'])
print(f"Dropped notes column")

# Drop rows where date is still null - unrecoverable
before = bw.shape[0]
bw = bw.dropna(subset=['date'])
print(f"Dropped {before - bw.shape[0]:,} rows with unrecoverable dates")

# Drop rows where amount is null - can't model without it
before = bw.shape[0]
bw = bw.dropna(subset=['amount'])
print(f"Dropped {before - bw.shape[0]:,} rows with null amounts")

# Fill missing category with 'Unknown'
bw['category'] = bw['category'].fillna('Unknown')

# Fill missing payment_mode with mode (most common value)
bw['payment_mode'] = bw['payment_mode'].fillna(bw['payment_mode'].mode()[0])

# Fill missing location with 'Unknown'
bw['location'] = bw['location'].fillna('Unknown')

# Final check
print()
print(f"Final shape: {bw.shape[0]:,} rows x {bw.shape[1]} columns")
print(f"Missing values remaining:")
print(bw.isnull().sum())
```

```python
# Cell 14 - Save Cleaned Dataset
import os

# Create processed folder inside Data
os.makedirs('../Data/processed/', exist_ok=True)

# Save cleaned BudgetWise dataset
bw.to_csv('../Data/processed/budgetwise_cleaned.csv', index=False)

print("Cleaned dataset saved to Data/processed/budgetwise_cleaned.csv")
print(f"Final shape: {bw.shape[0]:,} rows x {bw.shape[1]} columns")
print()
print("Preview of cleaned data:")
print(bw.head())
```

```python
# Currency Conversion - Production Grade Implementation
import requests
from datetime import date
from dotenv import load_dotenv
import os

load_dotenv()

# Step 1 - Pull live exchange rates
def get_live_rates():
    api_key = os.getenv('FX_API_KEY')
    url = f"https://v6.exchangerate-api.com/v6/{api_key}/latest/USD"
    response = requests.get(url)
    data = response.json()
    
    if data['result'] == 'success':
        print(f"Exchange rates fetched successfully")
        print(f"Last updated: {data['time_last_update_utc']}")
        return data['conversion_rates']
    else:
        raise Exception(f"FX API error: {data['error-type']}")

rates = get_live_rates()
print(f"\nKey rates today ({date.today()}):")
print(f"  1 USD = {rates['INR']:.2f} INR")
print(f"  1 USD = {rates['EUR']:.4f} EUR")
print(f"  1 USD = {rates['GBP']:.4f} GBP")

# Step 2 - Detect original currency from raw amount
def detect_currency(value):
    value = str(value).strip()
    if value.startswith('₹'):
        return 'INR'
    elif value.startswith('$'):
        return 'USD'
    elif value.startswith('€'):
        return 'EUR'
    elif value.startswith('£'):
        return 'GBP'
    else:
        return 'USD'  # default to USD if no symbol

# Step 3 - Convert amount to USD
def convert_to_usd(amount_str, currency, rates):
    clean = (str(amount_str)
             .replace('₹', '')
             .replace('$', '')
             .replace('€', '')
             .replace('£', '')
             .replace(',', '')
             .strip())
    try:
        amount = float(clean)
    except:
        return None
    
    if currency == 'USD':
        return round(amount, 2)
    else:
        return round(amount / rates[currency], 2)

# Step 4 - Apply to BudgetWise dirty dataset
print("\nApplying currency conversion to BudgetWise dataset...")

# Reload fresh copy of dirty data
bw_currency = bw_dirty.drop_duplicates().copy()

# Detect currency for each transaction
bw_currency['original_currency'] = bw_dirty['amount'].apply(detect_currency)

# Show currency distribution
print("\nCurrency distribution in raw data:")
print(bw_currency['original_currency'].value_counts())

# Convert all amounts to USD
bw_currency['amount_usd'] = bw_currency.apply(
    lambda row: convert_to_usd(
        row['amount'],
        row['original_currency'],
        rates
    ), axis=1
)

# Store conversion metadata for auditability
bw_currency['exchange_rate_date'] = date.today().isoformat()
bw_currency['exchange_rate_used'] = bw_currency['original_currency'].map({
    'USD': 1.0,
    'INR': round(1/rates['INR'], 6),
    'EUR': round(1/rates['EUR'], 6),
    'GBP': round(1/rates['GBP'], 6)
})

# Step 5 - Show before and after comparison
print("\nBefore vs After conversion sample:")
sample = bw_currency[bw_currency['original_currency'] == 'INR'].head(5)
print(sample[['amount', 'original_currency', 
              'amount_usd', 'exchange_rate_used', 
              'exchange_rate_date']])

print(f"\nConversion complete:")
print(f"  USD transactions: {(bw_currency['original_currency']=='USD').sum():,}")
print(f"  INR transactions converted: {(bw_currency['original_currency']=='INR').sum():,}")
print(f"  Exchange rate used: 1 USD = {rates['INR']:.2f} INR")
print(f"  Rate date: {date.today()}")
```

```python
# Save currency-converted dataset
os.makedirs('../Data/processed/', exist_ok=True)

# Apply all previous cleaning steps to bw_currency
from dateutil import parser as dateutil_parser

def parse_date_flexible(date_str):
    if pd.isnull(date_str):
        return pd.NaT
    try:
        return dateutil_parser.parse(str(date_str), dayfirst=False)
    except:
        return pd.NaT

# Clean the currency converted dataset
bw_currency['date'] = bw_currency['date'].apply(parse_date_flexible)
bw_currency = bw_currency.dropna(subset=['date', 'amount_usd'])
bw_currency = bw_currency.drop(columns=['notes', 'amount'])
bw_currency = bw_currency.rename(columns={'amount_usd': 'amount'})
bw_currency['category'] = bw_currency['category'].fillna('Unknown')
bw_currency['payment_mode'] = bw_currency['payment_mode'].fillna(
    bw_currency['payment_mode'].mode()[0])
bw_currency['location'] = bw_currency['location'].fillna('Unknown')

# Apply fuzzy category cleaning
from thefuzz import process

valid_categories = [
    'Food', 'Rent', 'Travel', 'Utilities', 'Entertainment',
    'Salary', 'Bonus', 'Savings', 'Education', 'Others',
    'Other Income', 'Unknown', 'Healthcare'
]

def clean_category(cat):
    if pd.isnull(cat):
        return 'Unknown'
    match, score = process.extractOne(cat, valid_categories)
    if score >= 70:
        return match
    else:
        return 'Others'

bw_currency['category'] = bw_currency['category'].apply(clean_category)

# Save
bw_currency.to_csv('../Data/processed/budgetwise_usd_converted.csv', 
                   index=False)

print("Currency converted dataset saved")
print(f"Final shape: {bw_currency.shape[0]:,} rows x {bw_currency.shape[1]} columns")
print(f"Columns: {list(bw_currency.columns)}")
print()
print("Sample of final dataset:")
print(bw_currency.head())
```
